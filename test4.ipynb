```python
def owner_document():
    document_number = input('Введите номер документа: ')
    owner = ', '.join([owner['name'] for owner in documents if owner['number'] == document_number])

    if owner:
      print(f'Владелец документа: {owner}')
    else:
      print('Документа не существует')
```

```python
def document_shelf():
    document_number = input('Введите номер документа: ')
    shelf = next((shelf_num for shelf_num, document in directories.items() if document_number in document), None)

    if shelf:
      print(f'Документ хранится на полке: {shelf}')
    else:
      print('Документ не найден в базе')
    return shelf
```

```python
#Вспомогательная функция, возвращает номер полки документа
def number_for_all(document_num):
  shelf = next((shelf_num for shelf_num, document in directories.items() if document_num in document), None)
  return shelf

def all_info():
  for index, info in enumerate(documents):
    print(f"{index}: {info['number']}, тип: {info['type']}, владелец: {info['name']}, полка хранения: {number_for_all(info['number'])}")
```

```python
def add_shelf():
  number = input('Введите номер полки: ')

  if number in directories:
    print(f'Такая полка уже существует. Текущий перечень полок: {", ".join(list(directories.keys()))}')
  else:
    directories.setdefault(number, [])
    print(f'Полка добавлена. Текущий перечень полок: {", ".join(list(directories.keys()))}')

  # Еще один вариант, просто для интереса
  # if number not in directories:
  #   directories.setdefault(number, [])
  #   print(f'Полка добавлена. Текущий перечень полок: {", ".join([key for key in directories.keys()])}')
  #   return
  # print(f'Такая полка уже существует. Текущий перечень полок: {", ".join([key for key in directories.keys()])}')
```

```python
def del_shelf():
  number = input('Введите номер полки: ')

  if number not in directories:
    print(f'Такой полки не существует. Текущий перечень полок: {", ".join(list(directories.keys()))}.')
  elif directories[number]:
    print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {", ".join(list(directories.keys()))}.')
  else:
    del(directories[number])
    print(f'Полка удалена. Текущий перечень полок: {", ".join(list(directories.keys()))}.')
```

```python
def add_document():
  new_set = {
      'type': input('Введите тип документа: '),
      'number': input('Введите номер документа: '),
      'name': input('Введите владельца документа: ')
  }
  shelf_num = input('Введите полку для хранения: ')

  if shelf_num in directories:
    documents.append(new_set)
    directories[shelf_num].append(new_set['number'])
    print('Документ добавлен. Текущий список документов: ')
  else:
    print('Такой полки не существует. Добавьте полку командой as.\nТекущий список документов: ')
  all_info()
```

```python
#Вспомогательная функция для удаления документа из полки. Не знаю как сделать ее лаконичнее
def del_shelf_doc(document_number):
  for num, values in directories.items():
    for val in values:
      if document_number in val:
        directories[num].remove(val)

def del_document():
  number = input('Введите номер документа: ')
  found = False
  i = 0

  for doc in documents:
    if doc['number'] == number:
      del(documents[i])
      found = True
      del_shelf_doc(number)
      print('Документ удален.\nТекущий список документов:')
    i += 1
  if not found:
    print('Документ не найден в базе.\nТекущий список документов:')

  all_info()
```

```python
def replace_doc():
  number = input('Введите номер документа: ')
  shelf = input('Введите номер полки: ')
  document_found = False

  if shelf not in directories.keys():
    print(f'Такой полки не существует. Текущий перечень полок: {", ".join(list(directories.keys()))}.')
    return

  for values in directories.values():
    if number in values:
      document_found = True
      break

  if not document_found:
    print('Документ не найден в базе.\nТекущий список документов:')
    all_info()
    return

  del_shelf_doc(number)
  directories[shelf].append(number)
  print('Документ перемещен.\nТекущий список документов:')
  all_info()
```

```python
def main():

  while True:
    command = input('Введите команду: ')
    match command:
      case 'q':
        print('Завершение программы')
        break
      case 'p':
        owner_document()
      case 's':
        document_shelf()
      case 'l':
        all_info()
      case 'ads':
        add_shelf()
      case 'ds':
        del_shelf()
      case 'ad':
        add_document()
      case 'd':
        del_document()
      case 'm':
        replace_doc()

main()
```
